```python
# 1. 데이터 불러오기
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
# 4. 모델 학습 및 성능 비교 (기본 모델부터 다양한 모델까지)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# 7. 모델 저장 및 로드
import joblib
```

```python
# 파일 경로 설정
train_path = 'dataset/train.csv'
test_path = 'dataset/test.csv'

# 데이터 불러오기
train_df = pd.read_csv(train_path, index_col='ID')
test_df = pd.read_csv(test_path, index_col='ID')
```

```python
# 각 열의 데이터 타입을 확인
print(train_df.dtypes)

# 숫자형(Numerical) 열만 선택
numerical_columns = train_df.select_dtypes(include=['int64', 'float64']).columns
print("Numerical columns:", numerical_columns)

# 범주형(Categorical) 열만 선택
categorical_columns = train_df.select_dtypes(include=['object', 'category']).columns
print("Categorical columns:", categorical_columns)
```

```python
# 타겟 값 분리
X = train_df.drop(columns=['SUBCLASS'])
y = train_df['SUBCLASS']

# 2. 인코딩 (Label Encoding 사용)
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

# X의 모든 열이 범주형이라고 가정하고, Label Encoding 적용
for col in X.columns:
    X[col] = label_encoder.fit_transform(X[col])

# test 데이터도 동일한 인코딩 처리
for col in test_df.columns:
    test_df[col] = label_encoder.fit_transform(test_df[col])
```

```python
# 3. Train-Test Split (stratify 사용)
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
# 모델 리스트
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Random Forest': RandomForestClassifier(),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='mlogloss'),
    "SupportVectorMachine": SVC(),
    "KNeighbors": KNeighborsClassifier(),
    "DecisionTree": DecisionTreeClassifier()
}
```

```python
# 모델 학습 및 성능 출력
for name, model in models.items():
    model.fit(X_train, y_train)

    # 모델 저장
    joblib.dump(model, './trained_model/'+name+'.pkl')

    y_pred = model.predict(X_valid)
    acc = accuracy_score(y_valid, y_pred)
    print(f'{name} Accuracy: {acc:.4f}')
```

```python
# 5. 앙상블 및 스태킹 모델 구현
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import StackingClassifier

# 앙상블 모델 (VotingClassifier 사용)
ensemble_model = VotingClassifier(estimators=[
    ('lr', LogisticRegression(max_iter=1000)),
    ('rf', RandomForestClassifier()),
    ('xgb', XGBClassifier(use_label_encoder=False, eval_metric='mlogloss'))
], voting='hard')

ensemble_model.fit(X_train, y_train)

# 모델 저장
joblib.dump(ensemble_model, './trained_model/ensemble.pkl')

y_pred_ensemble = ensemble_model.predict(X_valid)
ensemble_acc = accuracy_score(y_valid, y_pred_ensemble)
print(f'Ensemble Model Accuracy: {ensemble_acc:.4f}')
```

```python
# 스태킹 모델
stacking_model = StackingClassifier(estimators=[
    ('lr', LogisticRegression(max_iter=1000)),
    ('rf', RandomForestClassifier()),
    ('xgb', XGBClassifier(use_label_encoder=False, eval_metric='mlogloss'))
], final_estimator=LogisticRegression())

stacking_model.fit(X_train, y_train)

# 모델 저장
joblib.dump(stacking_model, './trained_model/stacking.pkl')

y_pred_stacking = stacking_model.predict(X_valid)
stacking_acc = accuracy_score(y_valid, y_pred_stacking)
print(f'Stacking Model Accuracy: {stacking_acc:.4f}')
```

```python
# 예측 결과를 Label Encoder로 원래 SUBCLASS 형식으로 변환
test_pred_original = label_encoder.inverse_transform(test_pred)

# submission 파일 생성
submission = pd.DataFrame({'ID': test_df.index, 'SUBCLASS': test_pred_original})
submission.to_csv('submission.csv', index=False)
```
